```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('RestaurantAnalysis.csv')
data
```

```python
data.info()
```

```python
plt.hist(data['City'],bins=14)
```

```python
sns.countplot(x=data['rate'],)
```

```python
sns.countplot(x=data['Type'])
```

```python
plt.hist(data['Approx_Cost'],bins=10)
```

```python
a=(data['City'])
b=(data['Type'])
plt.bar(a,b, label='C', linewidth=2)
plt.legend(loc='upper right')
```

```python
groupedby=data.groupby('City')['Approx_Cost'].sum()
result=pd.DataFrame({'Approx_Cost':groupedby})
plt.plot(result,c='red',marker='o')
```

```python
gbby=data.groupby('Name')['rate'].sum()
result=pd.DataFrame({'rate':gbby})
plt.plot(result,c='red')
plt.xlabel('Names',size=18,c='red')
plt.ylabel('Rate',size=18,c='red')
```

```python
plt.bar(data['Type'],data['Approx_Cost'],label='Restaurant Price according to type')
plt.legend()
```

```python
a=(data['Type'])
b=(data['City'])
c=(data['rate'])
plt.plot(b,c,label='Cities and votes',linewidth=2)
plt.plot(a,c,label='Types and votes',linewidth=2)
plt.legend('upperright')
```
